```python
import numpy as np
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import os
```

## Description
Scrapes all events and their attendees from the Affiliate Summit West 2025 Website.

## Logic
* Create array to hold events (events).
* Launch Selenium client at https://affiliatesummit.app.swapcard.com/event/affiliate-summit-west-2025/my-visit/my-schedule
* Manually login using Selenium + one time passcode.
* Create events_links list.
* For each event (link):
    - Create temporary list (event).
    - Selenium click() the event.
    - Selenium click ("see all").
    - For Each user:
        - Create dict (users) containing information: event_name, event_date, event_location, user_name, job_title, company_name, profile_link.
        - Append dict to event list.
* np.vstack event onto events.

## Events List

```python
# Events list
events_list = []
```

```python
# Create Soup
soup = BeautifulSoup(driver.page_source, "html.parser")
```

```python
# Events section
events_section = soup.find_all("div", class_="sc-de1a6ac-0 jbSazf")[1]
```

```python
# Links
links_dirty = events_section.find_all("a", href=True)
```

```python
# Instantiate Event Links
event_links = []
```

```python
# Iterate through links_dirty create list of event_links
partial_link = "https://affiliatesummit.app.swapcard.com"
for i in range(0, len(links_dirty)):
    event_link = partial_link + str(links_dirty[i]).split('"')[1]
    print(event_link)
    event_links.append(event_link)
```

## Iterate Through Event Links, Click on Each Event

```python
def scrape_event(link):
    # Go to the event
    driver.get(link)
    time.sleep(1)

    # Scroll to bottom of page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Click "see all button"
    see_all_button = driver.find_element(By.XPATH, "//span[contains(text(), 'See all')]")
    see_all_button.click()

    # Get soup
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Scrape event information
    event_name = soup.find("h3", class_="sc-57bc748-3 gjVHfO").text
    event_time = soup.find("div", class_="sc-901e7a18-2 eEIcVK").text
    event_location = soup.find("div", class_="sc-901e7a18-2 eEIcVK").text

    # Scroll 5 times
    div_to_scroll = driver.find_element(By.CLASS_NAME, "sc-143ccdb4-0.scyzj")
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", div_to_scroll)
    time.sleep(1)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", div_to_scroll)
    time.sleep(1)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", div_to_scroll)
    time.sleep(1)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", div_to_scroll)
    time.sleep(1)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", div_to_scroll)

    # Scrape User information
    user_names = [x.text for x in soup.find_all("span", class_="sc-a13c392f-0 sc-b2e9f250-9 ffVsRx iYQpoN")]
    user_job_titles = [x.text for x in soup.find_all("span", class_="sc-a13c392f-0 sc-b2e9f250-11 ffVsRx gTTUUz")]
    user_job_names = [x.text for x in soup.find_all("span", class_="sc-a13c392f-0 sc-b2e9f250-12 ffVsRx fJibeY")]

    # Create dataframe
    events_dict = {"event_name": [event_name] * len(user_names),
                   "event_time": [event_time] * len(user_names),
                   "event_location": [event_location] * len(user_names),
                   "user_name": user_names,
                   "user_job_titles": user_job_titles,
                   "user_company_names": user_job_names}

    return event_name, events_dict
```

```python
driver.find_element(By.XPATH, "//span[contains(text(), 'See all')]").click()
```

#### Test on first event

```python
all_event_data = {}

# Iterate through event links
for link in event_links[0:1]:
    # Get the event name and event dict
    event_name, events_dict = scrape_event(link)

    # Convert events dict to dataframe, store it in events_dict
    all_event_data[event_name] = pd.DataFrame(events_dict)
```

#### Run on all events

```python
all_event_data = {}

# Iterate through event links
for link in event_links:
    # Get the event name and event dict
    event_name, events_dict = scrape_event(link)

    # Convert events dict to dataframe, store it in events_dict
    all_event_data[event_name] = pd.DataFrame(events_dict)
```

```python
# Save directory
save_directory = "event_data"

# Save each dataframe as a CSV
for event_name, df in all_event_data.items():
    # Create a safe filename by replacing problematic characters
    safe_event_name = event_name.replace("/", "-").replace("?", "").replace(":", "").replace("|", "").replace("*", "").replace('"', '').replace('<', '').replace('>', '')
    
    # Define full path
    file_path = os.path.join(save_directory, f"{safe_event_name}.csv")
    
    # Save dataframe to CSV
    df.to_csv(file_path, index=False)

# Confirm saved files
saved_files = os.listdir(save_directory)
saved_files
```
