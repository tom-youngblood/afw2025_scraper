# Merge all Leads to Create Final CSV

```python
import pandas as pd
```

## Glimpse CSVs

```python
leads = pd.read_csv("leads_links.csv")
enr = pd.read_csv("all_profiles_final.csv")
```

```python
leads.head()
```

## Perform Merge

```python
df = pd.merge(left=leads, right=enr, how="inner", left_on="Full Link", right_on="Profile URL")
```

```python
df = df.drop(columns=["Unnamed: 0", "Profile Link"])
```

```python
df
```

```python
df.to_csv("lead_data.csv")
```
